##Pergunta:
O que é um descritor em visão computacional?


**Resposta:**

Um descritor em visão computacional é uma representação numérica ou vetorizada que captura características distintivas de uma imagem ou região de interesse, permitindo comparações, correspondências e reconhecimento de padrões.

##Pergunta:
Quais são as vantagens do algoritmo SIFT em visão computacional?

**Resposta:**

O algoritmo SIFT apresenta diversas vantagens. Primeiro, ele é robusto a mudanças na escala, rotação, iluminação e até mesmo parcialidade de objetos. Além disso, as características extraídas pelo SIFT são invariantes a transformações afins, o que os torna adequados para uma ampla gama de aplicações. O SIFT também é capaz de lidar com oclusões parciais e variações de aparência, tornando-o adequado para tarefas como reconhecimento de objetos e correspondência de imagens.

##Pergunta:
Como as características extraídas por extratores de características robustos são invariáveis a rotação, escala e translação?

**RESPOSTA**

A invariância é alcançada através de uma série de transformações aplicadas nas características extraídas. A invariância à escala é geralmente alcançada identificando a característica em diferentes escalas de uma imagem, que é feito através da aplicação repetida de um filtro gaussiano para suavizar a fotografia, e depois reduzindo a resolução da imagem downsampling para criar uma pirâmide.

Na invariância à rotação, após os pontos de interesse serem detectados, uma orientação é atribuída a cada ponto, que por si só é invariante à rotação porque a descrição de um ponto de interesse pode ser rotacionada para corresponder à sua orientação antes de ser comparada com outros descritores.

A translação é a mudança na localização de um objeto. Como os descritores de características são baseados na aparência local de um objeto (por exemplo, a distribuição de gradientes de intensidade), eles não são afetados pela translação. Se um objeto se move para uma nova localização na imagem, a aparência local permanecerá a mesma, portanto, os descritores de características também permanecerão os mesos.

##Exercício Prático:
O Instituto de Informática da UFG deseja treinar algoritmos de Inteligência Artificial para reconhecimento de imagens de sua fachada, sua missão é implementar um algoritmo para extrair características de uma imagem da fachada do INF utilizando o descritor SIFT e assim ajudar na geração de features para os modelos que serão criados.

**Solução**

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
!gdown 1T6UwGY6i2mKjVSE2AQlzcS3GoWDonCUK
image = cv2.imread('inf_ufg.jpeg')
```

```python
sift = cv2.SIFT_create() # criando objeto de extração de características SIFT
keypoints, descriptors = sift.detectAndCompute(image, None) # detectando e extraindo características
```

```python
image_with_keypoints = cv2.drawKeypoints(image, keypoints, None) # desenhando os keypoints nas imagens

# exibindo imagem e características extraídas
cv2_imshow(image_with_keypoints)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
